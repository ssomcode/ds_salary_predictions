```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('eda_data.csv')
```

```python
df.head()
```

```python
df.describe().transpose()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
#steps involved:
#1.chose relevent column
#2.get dummy variables
#3.train test split
#4.linear regression
#5.lasso regression
#6.random forest
#7. tune models using gridSearchCV
#8.test ensambles
```

```python
#1. chose relevent columns

df.columns
```

```python
df_model=df[['avg_sal','Rating','Size','Type of ownership', 'Industry', 'Sector', 'Revenue','hourly','employer_provided',
       'company_text', 'job_state', 'same_state', 'comapny_age', 'python','spark', 'aws', 'excel', 'title_simp', 'seniority',
       'desc_len','num_comp' ]]
```

```python
df_dum=pd.get_dummies(df_model,drop_first=True)
```

```python
df_dum.head()
```

```python
#train test split 
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X=df_dum.drop('avg_sal',axis=1)
y=df_dum['avg_sal'].values
```

```python
X_train,X_test,y_train,y_test=tts(X,y,random_state=101,test_size=0.3)
```

```python
import statsmodels.api as sm
```

```python
X_sm = X = sm.add_constant(X)
```

```python
model=sm.OLS(y,X_sm)
```

```python
model.fit().summary()
```

```python
from sklearn.linear_model import LinearRegression as lr
from sklearn.model_selection import cross_val_score
```

```python
lm=lr()
```

```python
lm.fit(X_train,y_train)
```

```python
cross_val_score(lm,X_train,y_train,scoring='neg_mean_absolute_error',cv=3)
```

```python
#this tell us how far off we are from original values
```

```python
np.mean(cross_val_score(lm,X_train,y_train,scoring='neg_mean_absolute_error',cv=3))
```

```python
#lasso regression
```

```python
from sklearn.linear_model import Lasso
```

```python
lm_l=Lasso()
```

```python
lm_l.fit(X_train,y_train)
```

```python
np.mean(cross_val_score(lm_l,X_train,y_train, scoring = 'neg_mean_absolute_error', cv= 3))
```

```python
# lets make a loop to select better alpha
```

```python
alpha=[]
error=[]


```

```python
for i in range(1,100):
    alpha.append(i/100)
    lml=Lasso(alpha=i/100)
    error.append(np.mean(cross_val_score(lml,X_train,y_train,scoring='neg_mean_absolute_error',cv=3)))

plt.plot(alpha,error)
```

```python
err = list(zip(alpha,error))           #zip will iterate theough first list and second list and pair respective elements  
df_err = pd.DataFrame(err, columns = ['alpha','error'])
df_err[df_err.error == max(df_err.error)]
```

```python
lm_l=Lasso(alpha=0.02)
```

```python
lm_l.fit(X_train,y_train)
```

```python
np.mean(cross_val_score(lm_l,X_train,y_train, scoring = 'neg_mean_absolute_error', cv= 3))
```

```python
# Random forest
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
rf.fit(X_train,y_train)
```

```python
np.mean(cross_val_score(rf,X_train,y_train,scoring='neg_mean_absolute_error',cv=3))
```

```python
#it has lesser error than lasso regression
```

```python
#tuning models with GridSearchCV
```

```python
#it takes all the parameters and outputs a model with best performance
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'n_estimators':range(10,300,10), 'criterion':('mse','mae'), 'max_features':('auto','sqrt','log2')}
```

```python
gs = GridSearchCV(rf,parameters,scoring='neg_mean_absolute_error',cv=3)
```

```python
gs.fit(X_train,y_train)
```

```python
gs.best_score_
```

```python
gs.best_estimator_
```

```python
#so best model was random forest with score of -15.4642
```

```python
#lets test the test data
```

```python
tpred_lm=lm.predict(X_test)
tpred_lml=lm_l.predict(X_test)
tpred_rf=rf.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(mean_absolute_error(y_test,tpred_lm))
```

```python
print(mean_absolute_error(y_test,tpred_lml))
```

```python
print(mean_absolute_error(y_test,tpred_rf))
```

```python
#lets see if their mean is good or not 
```

```python
(tpred_lml+tpred_rf)/2
```

```python
#so it worse than the random forest
```

```python
print(mean_squared_error(y_test,tpred_rf))
```
