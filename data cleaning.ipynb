Now our task is to reomve all text and paranthesis content from the salary estimate column
we can do this by using lambda function
first we will split the expression at the paranthesis

# salary parsing

```python
df['hourly'] = df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
df['employer_provided'] = df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary:' in x.lower() else 0)
```

```python
salary=df['Salary Estimate'].apply(lambda x:x.split("(")[0])
```

```python
salary
```

```python
# now we will replace $ and k with blank
```

```python
rep_k=salary.apply(lambda x:x.replace('K',''))
```

```python
rep_k
```

```python
rep_d=rep_k.apply(lambda x:x.replace('$',''))
```

```python
rep_d
```

```python
#now we will make two different columns by hourly salary and employer provided salary
```

```python
df.head(50)
```

```python
df['Employer_provided']=df['Salary Estimate'].apply(lambda x: 1 if "employer provided salary:" in x.lower() else 0)
```

```python
df.head(100)
```

```python
min_hr=rep_d.apply(lambda x:x.lower().replace("per hour",'').replace("employer provided salary:",''))
```

```python
min_hr
```

```python
#now the text has been removed
```

```python
#Now we will find the average salary 
```

```python
df['min_sal']=min_hr.apply(lambda x:x.split('-')[0])
df['min_sal']
```

```python
df['max_sal']=min_hr.apply(lambda x:x.split('-')[1])
df['max_sal']
```

```python
df['min_sal'].to_int
```

```python
df['min_sal'] = pd.to_numeric(df['min_sal'])
#df['DataFrame Column'] = pd.to_numeric(df['DataFrame Column'])
```

```python
df['min_sal']
```

```python
df['max_sal']=pd.to_numeric(df['max_sal'])
```

```python
df['max_sal']
```

```python
df['avg_sal']=(df.min_sal+df.max_sal)/2
```

```python
df['avg_sal']
```
